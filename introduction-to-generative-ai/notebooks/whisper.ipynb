```python
import json
import os
import pandas as pd
import openai
from openai import OpenAI
```

```python
client = OpenAI()
```

```python
audio_file = open("audio.mp3", "rb")
```

```python
transcript = client.audio.transcriptions.create(
    file=audio_file,
    model="whisper-1",
    response_format="text"
)
```

```python
transcript.text
```

```python
system_prompt = "You are a proficient AI with a speciality in distilling information into key points. Based on the following text, identify and list the main points that we discussed or brought up. These should be the most important ideas, findings, or topics that are crucial to the essence of the discussion. Your goal is to a list that someone could read to quickly understand what was talked about."
```

```python
response = client.chat.completions.create(
    model="gpt-4-turbo",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": transcript.text}
    ],
    temperature=0.7,
    max_tokens=500
)
```
